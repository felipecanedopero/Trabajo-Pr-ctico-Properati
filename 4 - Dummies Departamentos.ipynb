```python
# Importamos las librerías que usaremos durante la limpieza:
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import re
import scipy.stats as sts
from bokeh.plotting import figure
from bokeh.io import output_notebook, show
from IPython.display import display
```

```python
df_properatti_depto = pd.read_csv("prop_apartament_1.csv")

from sklearn.feature_extraction.text import CountVectorizer
```

```python
#DUMMY BALCON

#Creo la primera columna para description
vec = CountVectorizer(strip_accents="unicode", lowercase=True, token_pattern="balcon", analyzer="word", binary=True)
X = vec.fit_transform(df_properatti_depto["description"])
dummy_balcon = pd.DataFrame(data = X.todense(), columns = vec.get_feature_names())
df_properatti_depto = pd.merge(df_properatti_depto, dummy_balcon, on = df_properatti_depto.index)
#Dropeo
df_properatti_depto.drop(["key_0"], axis = 1, inplace = True) 

#Creo la segunda para title
vec = CountVectorizer(strip_accents="unicode", lowercase=True, token_pattern="balcon", analyzer="word", binary=True)
X = vec.fit_transform(df_properatti_depto["title"])
dummy_balcon2 = pd.DataFrame(data = X.todense(), columns = vec.get_feature_names())
df_properatti_depto = pd.merge(df_properatti_depto, dummy_balcon2, on = df_properatti_depto.index)
#Dropeo
df_properatti_depto.drop(["key_0"], axis = 1, inplace = True)

#Junto las dos(la primera y la segunda) en una sola columna nueva
df_properatti_depto["balcon"] = df_properatti_depto["balcon_x"]|df_properatti_depto["balcon_y"]
##Dropeo las dos que use, dejando solo la nueva
df_properatti_depto.drop(["balcon_x"], axis = 1, inplace = True) 
df_properatti_depto.drop(["balcon_y"], axis = 1, inplace = True) 
```

```python
#DUMMY CENTRICO

#Creo la primera columna para description
vec = CountVectorizer(strip_accents="unicode", lowercase=True, token_pattern="(centrico|centro)", analyzer="word", binary=True)
X = vec.fit_transform(df_properatti_depto["description"])
dummy_centrico = pd.DataFrame(data = X.todense(), columns = vec.get_feature_names())
df_properatti_depto = pd.merge(df_properatti_depto, dummy_centrico, on = df_properatti_depto.index)
df_properatti_depto["centrico_x"] = df_properatti_depto["centrico"]|df_properatti_depto["centro"]
#Dropeo
df_properatti_depto.drop(["key_0"], axis = 1, inplace = True)
df_properatti_depto.drop(["centrico"], axis = 1, inplace = True) 
df_properatti_depto.drop(["centro"], axis = 1, inplace = True) 

#Creo la segunda columna para title
vec = CountVectorizer(strip_accents="unicode", lowercase=True, token_pattern="(centrico|centro)", analyzer="word", binary=True)
X = vec.fit_transform(df_properatti_depto["title"])
dummy_centrico2 = pd.DataFrame(data = X.todense(), columns = vec.get_feature_names())
df_properatti_depto = pd.merge(df_properatti_depto, dummy_centrico2, on = df_properatti_depto.index)
df_properatti_depto["centrico_y"] = df_properatti_depto["centrico"]|df_properatti_depto["centro"]
#Dropeo
df_properatti_depto.drop(["key_0"], axis = 1, inplace = True)
df_properatti_depto.drop(["centrico"], axis = 1, inplace = True) 
df_properatti_depto.drop(["centro"], axis = 1, inplace = True)

#Junto las dos(la primera y la segunda) en una sola columna nueva, en la que habra un 1 si se encontró coincidencia en al menos una de las columnas de las anteriores) 
df_properatti_depto["centrico"] = df_properatti_depto["centrico_x"]|df_properatti_depto["centrico_y"]
#Dropeo las dos que use, dejando solo la nueva(en la cual busque las palabras frente y vista al mar en description y en title)
df_properatti_depto.drop(["centrico_x"], axis = 1, inplace = True) 
df_properatti_depto.drop(["centrico_y"], axis = 1, inplace = True) 
```
